```python
import numpy as np
from sklearn.model_selection import train_test_split
import pandas as pd

dfdec = pd.read_csv("Tweets_Dec.csv")

train_dec, test_dec = train_test_split(dfdec, test_size=100, random_state=200)

train_dec.to_csv('Train_Dec.csv',index=False)

test_dec.to_csv('Test_Dec.csv',index=False)

dffeb = pd.read_csv("Tweets_Feb.csv")

train_feb, test_feb = train_test_split(dffeb, test_size=100, random_state=200)

train_feb.to_csv('Train_Feb.csv',index=False)

test_feb.to_csv('Test_Feb.csv',index=False)
```
